```python
# List of specific columns to drop based on type and redundancy
columns_to_drop = [
    'AgencyName_x', 'ScheduleSlotId', 'ClientId', 'CaregiverId',
    'Client Name', 'Caregiver Name',
    'AgencyId_x', 'AgencyName_y', 'AgencyId_y', 'Start Date', 'End Date',
    'FirstName', 'LastName', 'Address1', 'Address2',
    'SeparationNotes', 'Created Date', 'AISendDate', 'modifiedon',
    'Client State', 'Schedule Status', 'StateName', 'CountryName',
    'CaregiverRole', 'IsActive', 'IsArchived', 'ModifiedOn',
    'ZipCodes', 'CityName', 'Client City', 'Client Gender',
    'Bill Unit Number', 'Pay Unit Number', 'Caregiver Status',
    'Caregiver Restriction', 'Skille Type', 'Termination Date'
]

# Drop columns with more than 80% missing values
missing_threshold = 0.8
high_missing_cols = df.columns[df.isnull().mean() > missing_threshold].tolist()

# Combine all columns to drop
all_columns_to_drop = list(set(columns_to_drop + high_missing_cols))

# Drop the columns
df_cleaned = df.drop(columns=all_columns_to_drop)

# Show remaining columns
print(f"Remaining columns after dropping: {df_cleaned.shape[1]}")
print(df_cleaned.columns.tolist())
```

```python
# Separate numerical and categorical columns
numerical_cols = df_cleaned.select_dtypes(include=['number']).columns
categorical_cols = df_cleaned.select_dtypes(include=['object', 'category']).columns

# Fill numerical columns with median
df_cleaned[numerical_cols] = df_cleaned[numerical_cols].fillna(df_cleaned[numerical_cols].median())

# Fill categorical columns with mode (most frequent value)
for col in categorical_cols:
    df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mode()[0])

# Check if any missing values remain
missing_summary = df_cleaned.isnull().sum()
missing_summary = missing_summary[missing_summary > 0]

if missing_summary.empty:
    print("All missing values have been handled.")
else:
    print("Remaining missing values:")
    print(missing_summary)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Prepare model and data
X = df_encoded.drop("IsCaregiverTerminated", axis=1)
y = df_encoded["IsCaregiverTerminated"]

# LightGBM model
lgb_model = LGBMClassifier(random_state=42)

# Stratified 5-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Store results
acc_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    lgb_model.fit(X_train, y_train)
    y_pred = lgb_model.predict(X_test)

    acc_scores.append(accuracy_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))

# Report average results
print(f"Average Accuracy: {np.mean(acc_scores):.4f}")
print(f"Average Precision: {np.mean(precision_scores):.4f}")
print(f"Average Recall: {np.mean(recall_scores):.4f}")
print(f"Average F1-score: {np.mean(f1_scores):.4f}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Updated evaluation function with confusion matrix
def evaluate_model_with_cm(name, model):
    acc, prec, rec, f1 = [], [], [], []
    cm_total = np.array([[0, 0], [0, 0]])

    for train_idx, test_idx in skf.split(X, y):
        X_train, X_val = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[test_idx]

        model.fit(X_train, y_train)
        y_pred = model.predict(X_val)

        acc.append(accuracy_score(y_val, y_pred))
        prec.append(precision_score(y_val, y_pred))
        rec.append(recall_score(y_val, y_pred))
        f1.append(f1_score(y_val, y_pred))

        # Add confusion matrix
        cm = confusion_matrix(y_val, y_pred)
        cm_total += cm

    # Plot confusion matrix
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm_total, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{name} - Confusion Matrix (5-Fold Total)')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    return {
        'Model': name,
        'Accuracy': np.mean(acc),
        'Precision': np.mean(prec),
        'Recall': np.mean(rec),
        'F1 Score': np.mean(f1)
    }

# Run for all models and display
results_cm = [evaluate_model_with_cm(name, model) for name, model in models.items()]

# Final metrics table
results_cm_df = pd.DataFrame(results_cm)
print(results_cm_df.sort_values(by='F1 Score', ascending=False))
```
